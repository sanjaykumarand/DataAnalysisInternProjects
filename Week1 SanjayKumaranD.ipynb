```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
df = pd.read_csv("n.csv")
df.shape
df.dtypes
df.head()
```

```python
Drop=df.drop(columns=['cast','director','description'])
print("Dropped Columns:\n",Drop)
df=Drop
```

```python
df=df.dropna()
df
```

```python
df=df.drop_duplicates()
df
```

```python
df['release_year'].astype(int)
```

```python

main_genre = df['show_id']
print("Extracted Column:\n",main_genre)
```

```python
df.shape
```

```python
import seaborn as sns
sns.countplot(x='type', data=df)
```

```python
sns.countplot(x='rating', data=df, order=df['rating'].value_counts().index)
```

```python
main_genre.value_counts().head(10).plot(kind='bar')
```

```python
df['release_year'].value_counts().sort_index().plot(kind='line')
```

```python
df['country'].value_counts().head(10).plot(kind='bar')
```

```python
movie_df = df[df['type'] == 'Movie'].copy()
movie_df['duration_min'] = movie_df['duration'].str.extract('(\d+)').astype(float)
sns.histplot(movie_df['duration_min'], bins=30, kde=True)
```

```python
tv_df = df[df['type'] == 'TV Show'].copy()
tv_df['season_count'] = tv_df['duration'].str.extract('(\d+)')[0].fillna(0).astype(int)
tv_df['season_group'] = tv_df['season_count'].apply(lambda x: str(x) if x <= 10 else '10+')
sns.countplot(x='season_group', data=tv_df, palette='pastel')
```
